```python
from rdkit.Chem import rdChemReactions
from agave_chem import MCSReactionMapper, NeuralReactionMapper, TemplateReactionMapper

neural_mapper = NeuralReactionMapper("neural_mapper")
mcs_mapper = MCSReactionMapper("mcs_mapper")
template_mapper = TemplateReactionMapper("template_mapper")
```

```python
result = neural_mapper.map_reaction("CC(Cl)(Cl)OC(C)(Cl)Cl.CC(=O)C(=O)O>>CC(=O)C(=O)Cl")
print(result["selected_mapping"])
rdChemReactions.ReactionFromSmarts(result["selected_mapping"], useSmiles=True)
```

```python
result = mcs_mapper.map_reaction("CC(Cl)(Cl)OC(C)(Cl)Cl.CC(=O)C(=O)O>>CC(=O)C(=O)Cl")
print(result["selected_mapping"])
rdChemReactions.ReactionFromSmarts(result["selected_mapping"], useSmiles=True)
```

```python
result = template_mapper.map_reaction("CC(Cl)(Cl)OC(C)(Cl)Cl.CC(=O)C(=O)O>>CC(=O)C(=O)Cl")
print(result["selected_mapping"])
rdChemReactions.ReactionFromSmarts(result["selected_mapping"], useSmiles=True)
```
